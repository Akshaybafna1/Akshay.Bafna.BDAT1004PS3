### Step 1 Insert the libraries


```python
#inserting the libraries which is step 1 as mentioned above 
import pandas as pd
import numpy as np
```

### Step 2 Import the dataset


```python
# importing the dataset from the prescribed address
DatasetLink = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
```

### Step 3 : Assign it to a variable called users

```python
#assigning this to a variable named users
users = pd.read_csv(DatasetLink , delimiter='|')
```

### Step 4: Discover what is the mean age per occupation

```python
#discovering the mean age per occupation from the data 
meanOfAge_per_occupation = users.groupby('occupation')['age'].mean()
print("meanOfAge_per_occupation:")
print(meanOfAge_per_occupation)
```

### Step 5: Discover the Male ratio per occupation and sort it from the most to the least

```python
#discoveringing the male ratio but not mentioning with the heading as male ratio just directly the result
male_ratio_by_occupation = users.groupby('occupation').apply(lambda x: (x['gender'] == 'M').mean()).sort_values(ascending=False)
male_ratio_by_occupation
```

### Step 6 : For each occupation, calculate the minimum and maximum ages

```python
#calculating the min and max age per occupation 
min_max_age_per_occupation = users.groupby('occupation')['age'].describe()[['min', 'max']]

min_max_age_per_occupation
```

### Step 7 : For each combination of occupation and sex, calculate the mean age

```python
#calculating the mean age for each combination of occupation and gender
mean_age_by_gender_and_occupation = users.groupby(['occupation', 'gender'])['age'].mean()
mean_age_by_gender_and_occupation
```

### Step 8 : For each occupation present the percentage of women and men

```python
# Group the data by occupation and gender, and count the number of users in each group
gender_counts = users.groupby(['occupation', 'gender']).size()

# Group the data by occupation and calculate the total number of users in each occupation
occupation_totals = users.groupby('occupation').size()

# Calculate the percentage of women and men for each occupation
women_per = gender_counts.loc[:, 'F'] / occupation_totals * 100
men_per = gender_counts.loc[:, 'M'] / occupation_totals * 100

# Create a new DataFrame with the results
gender_percentages = pd.DataFrame({'Percentage Women': women_per, 'Percentage Men': men_per})

# Display the result
gender_percentages
```

```python
# assign variable and replace 
data = pd.read_csv(url,sep = '\s+',parse_dates=[[0,1,2]])

data = data.rename(columns={'Yr_Mo_Dy': 'DATE'})


data['DATE'] = pd.to_datetime(data['DATE'], format='%Y_%m_%d')
data = data.set_index('DATE')

data.head()
```

### Step 4 : Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
# Define a function to adjust year
def fix_year(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return pd.to_datetime(year * 10000 + x.month * 100 + x.day, format='%Y%m%d')

# Apply the function to fix the year in the date column
data.index = data.index.map(fix_year)

data.head()
```

```python
# Read the data 
url = 'us-marriages-divorces-1867-2014.csv'
US_MRG_DIV = pd.read_csv(url)

# Set the year column as the index
US_MRG_DIV.set_index('Year', inplace=True)

# Create a line plot showing the number of marriages and divorces per capita
plt.plot(US_MRG_DIV['Marriages_per_1000'], label='Marriages')
plt.plot(US_MRG_DIV['Divorces_per_1000'], label='Divorces')

# Add labels and legend
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.xlabel('Year')
plt.ylabel('Number per 1000')
plt.legend()

# Show the plot
plt.show()
```

```python
#import lib
import pandas as pd
import matplotlib.pyplot as plt

# import the dataset
url = 'us-marriages-divorces-1867-2014.csv'
data = pd.read_csv(url)

#filter years
years = [1900, 1950, 2000]
df = data[data.Year.isin(years)]

#plotting data 
fig, ax = plt.subplots(figsize=(10, 8))
ax.bar(df.Year, df.Marriages_per_1000, width=2, align='edge', label='Marriages per capita')
ax.bar(df.Year, df.Divorces_per_1000, width=-2, align='edge', label='Divorces per capita')


ax.set_xlabel('Year')
ax.set_ylabel('Per capita')
ax.set_title('Marriages and Divorces per capita in the U.S. in 1900, 1950, and 2000')
ax.legend()
plt.show()
```

## Question 9

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('roman-emperor-reigns.csv')

# Count the occurrences of 'Assassinated' in the 'Cause_of_Death' column
assassinated_count = data['Cause_of_Death'].value_counts()['Assassinated']

# Count the total number of emperors
total_emperors = len(data)

# Calculate the percentage of emperors assassinated
percentage_assassinated = (assassinated_count / total_emperors) * 100

# Calculate the percentage of emperors who died from other causes
percentage_other = 100 - percentage_assassinated

# Create a pie chart with a different style
labels = ['Assassinated', 'Other Causes']
sizes = [percentage_assassinated, percentage_other]
colors = ['#ffcc99', '#c2c2f0']  # Different colors for a softer appearance
explode = (0.15, 0)  # Adjusted explode value for emphasis on 'Assassinated'

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140,
        wedgeprops=dict(width=0.4, edgecolor='w'))  # Added wedgeprops for a different appearance
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Fraction of Roman Emperors Assassinated')

# Add legend in the upper left corner
plt.legend(loc='upper left')

plt.show()
```
